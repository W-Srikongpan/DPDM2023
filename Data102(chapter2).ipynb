##Bar chart

## Stacked Barchart


````python
label = ['Oct', 'Nov', 'Dec','Jan','Feb','Mar','Apr','May','Jun']
plt.bar(range(len(table_frome_drive)),table_frome_drive['Amount_Sickleave'], color = 'b' , label = 'Sickleave')
plt.bar(range(len(table_frome_drive)),table_frome_drive['Amount_ Personalleave'],bottom=table_frome_drive['Amount_Sickleave'], color = 'r' , label = 'Personal leave')
plt.bar(range(len(table_frome_drive)),table_frome_drive['Amount_ Vacationleave'],bottom=table_frome_drive['Amount_Sickleave'] + table_frome_drive['Amount_ Personalleave'], color = 'g' , label = 'Vacation leave')
plt.xticks(range(len(table_frome_drive)),label)
plt.grid(True,axis='y')
plt.legend() ;
````

## grouped Barchart


````python
import numpy as np
width = 0.2
plt.bar(range(len(table_frome_drive)), table_frome_drive['Amount_Sickleave'],width,color='g',label='sick leave')
plt.bar(np.array(range(len(table_frome_drive)))-width, table_frome_drive['Amount_ Personalleave'],width,color='r',label='Personal leave')
plt.bar(np.array(range(len(table_frome_drive)))+width, table_frome_drive['Amount_ Vacationleave'],width,color='b',label='Vacation leave')
plt.legend();
````

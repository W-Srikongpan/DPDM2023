```python
transactions = data.groupby('CustomerID')['StockCode'].apply(list) #group by 'CustomerID' and 'StockCode'
transactions
```

#Association Rules

```python
#pip install apyori
from apyori import apriori
```

```python
frequentitemsets = list(apriori(transactions, min_support = 0.065))
#frequentitemsets
```

```python
len(frequentitemsets)
```

```python
min_sup = 0.03
frequentitemsets = list(apriori(transactions, min_support = min_sup))
for i in frequentitemsets:
    if i[0].__len__()>1:
        print('yes')
        print(i)
```
